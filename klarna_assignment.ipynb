#### User Defined Functions

```python
def drop_nan_col(data):
  '''
  This function takes a dataframe as input and drops the columns
  which has more than 40% of the data as NaN. It returns the new dataframe.
  '''
  row_count =data.shape[0] #total row count
  threshold = row_count*0.40 #defining threshold of 40% of total size
  counter =0 #counter flag

  for col in data:
    if col == 'default': #skipping the default column
      pass
    elif data[col].isnull().sum() > threshold : #dropping the column if has NaN more than threshold
      data= data.drop(columns=col, axis=1)
      counter +=1

  print('{} columns droped '.format(counter))
  return data #return the modified dataframe
```

```python
def fill_nan(data):
  '''
  This function takes dataframe as an input and 
  fill zero in place of NAN and returns the modified dataframe.
  '''
  
  for col in data.columns:
    if col == 'default': #skipping the default column
      pass
    elif data[col].isnull().sum()!=0: #checking if column has any NAN values
      data[col] = data[col].fillna(0) #fill those NAN with zero
      
  return data #return the modified data frame
```

```python
def convert_c2n(data):
  '''
  This function takes data frame as an input and converts 
  all the categorical or boolean features into numerical features,
  and returns the modified dataframe
  Here we are using Ordinal Encoder to execute the task.
  '''
  counter =0
  encoder = OrdinalEncoder() #initiallizing the encoder
  for col in data.columns:
    if data[col].dtype == 'O' or data[col].dtype == 'bool': #checking for object or boolean data type
      data[col]= encoder.fit_transform(data[col].values.reshape(-1,1))
      counter +=1
  print('{} columns converted to numerical '.format(counter))
  return data  #return the modified data frame
```

```python
def standard_scaling(data, test_flag = False):
  '''
  This function is used for normalising the data. 
  We use Standard scaler to to fit tranform the data 
  on the scale of [0,1] using z-score. 
  '''
  x_std = StandardScaler() #initialling the scaler
  if test_flag == False:
    # Standardizing the features, to bring all the featues on a scale between [0,1]
    x= x_std.fit_transform(data)
  else:
    x = x_std.fit(data)
  return x #return the modified values as an array
```

```python
def dimensionality_reduction(data, graph_flag=True):
  '''
  This function takes a data frame as input and apply PCA for dimentionality reduction
  and returns the components which has high variance and will have more feature importance 
  In this case we have initialized PCA for 95% variance.
  '''
  pca = PCA(0.95) # Initialising the PCA with variance of 95%
  principalComponents = pca.fit_transform(data)

  print('Explained Variance: ', pca.explained_variance_)
  print('\n')
  print('Count of PCA components: ', pca.n_components_)
  print('\n')
  print('Visulating the variance vs no.of components')

  if graph_flag ==True:
    # visulating the variance curve vs no.of components
    exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)
    fig = px.area(
        x=range(1, exp_var_cumul.shape[0] + 1),
        y=exp_var_cumul,
        labels={"x": "# Components", "y": "Explained Variance"}
    )
    fig.show()

  return principalComponents
```

```python
def correlation_heatmap(data):
  '''
  This function is to visualize a Correation matrix given a df.
  '''
  correlations = data.corr()
  fig, ax = plt.subplots(figsize=(25,25))
  sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
  plt.show();
```

```python
def plot_cm(labels, predictions, p=0.5):
  cm = confusion_matrix(labels, predictions > p)
  plt.figure(figsize=(5,5))
  sns.heatmap(cm, annot=True, fmt="d")
  plt.title('Confusion matrix @{:.2f}'.format(p))
  plt.ylabel('Actual label')
  plt.xlabel('Predicted label')

  precision = cm[1][1]/(cm[1][1] + cm[0][1])
  recall = cm[1][1]/(cm[1][1] + cm[1][0]) 
  f1 = 2 * ((precision * recall) / (precision + recall))
  print('(True Negatives): ', cm[0][0])
  print('(False Positives): ', cm[0][1])
  print('(False Negatives): ', cm[1][0])
  print('(True Positives): ', cm[1][1])
  print('Percision ' ,precision )
  print('Recall ' ,recall )

  # F1 which is a function of Precision and Recall. 
  # it is a good measure to use if we need to seek a balance between Precision and Recall 
  # and when there is an uneven class distribution
  print('f1 ' , f1)
```

```python
def print_auc(model):
  
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_val)
    
    fpr, tpr, threshold = roc_curve(y_train, y_train_pred)
    roc_auc = auc(fpr, tpr)
    print("Training AUC:", roc_auc)

    fpr, tpr, threshold = roc_curve(y_val, y_test_pred)
    roc_auc = auc(fpr, tpr)
    print("Val AUC:", roc_auc)
```

```python
def plot_metrics(history):
  metrics = ['loss', 'precision', 'recall' , 'f1']
  for i, metric in enumerate(metrics):
    name = metric.replace("_"," ").capitalize()
    
    plt.figure(figsize=(8, 8))

    plt.subplot(4,1,i+1 )
    
    plt.plot(history.epoch, 
             history.history[metric],
             color=colors[0], 
             label='Train')
    
    plt.plot(history.epoch,
             history.history['val_'+metric],
             color=colors[0], 
             linestyle="--",
             label='Val')
    
    plt.xlabel('Epoch')
    plt.ylabel(name)

    if metric == 'loss':
      plt.ylim([0, plt.ylim()[1]])
    else:
      plt.ylim([0,1])

    plt.legend()
```

#### Model 3: Neural Network

```python
def f1(y_true, y_pred):
    def recall(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + K.epsilon())
        return recall

    def precision(y_true, y_pred):
        true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
        predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + K.epsilon())
        return precision
    precision = precision(y_true, y_pred)

    recall = recall(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
metrics = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall'),
      tf.keras.metrics.AUC(name='auc'), # the probability that a classifier will rank  
      f1
]
```

```python
initial_bias = np.log([pos/neg])
initial_bias
```

```python
x =  np.asarray(principalComponents).astype('float32')
y = np.asarray(Y).astype('float32')

#Further spliting into train and val for testing purpose
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
def make_NN_model(metrics = metrics , *initialize_val , layers_1_size = 15):
  if initialize_val:
    initializer = tf.keras.initializers.Constant(initialize_val)
  else:
    initializer = None
    
  model = tf.keras.Sequential([
       tf.keras.layers.Dense(layers_1_size , activation= 'relu' ,
                             input_shape = (X_train.shape[1],) ),
      # to avoid overfitting we use dropout
       tf.keras.layers.Dropout(0.2),
       tf.keras.layers.Dense(1, activation='sigmoid',
                         bias_initializer=initializer)
      ]
  )
  model.compile(
      optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
      loss=tf.keras.losses.BinaryCrossentropy(),
      metrics=metrics)

  return model
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', 
    verbose=1,
    patience=5,
    mode='min',
    restore_best_weights=True)
batch_size = 2024
```

`batch_size = 2024`

Beacuase we want to make sure that each batch contains a few positives sample

If the batch size is too small, it would likely to have no defaulter attribute to learn from.
